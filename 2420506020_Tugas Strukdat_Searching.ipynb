```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
          high = mid - 1

    return -1

arr = [1000, 1005, 1010, 1020, 1030, 1070, 1080, 1200]
target = int(input("Masukkan Angka yang dicari: "))
result = binary_search(arr, target)

if result != -1:
    print(f"Elemen ditemukan pada indeks {result}")
else:
    print("Elemen tidak ditemukan")
```

```python
import random
import math

# 1. Membuat daftar 100 angka acak dalam rentang 1-1000 dan mengurutkannya
data = sorted(random.sample(range(1, 1001), 100))

# 2. Menentukan angka target (mengambil elemen tengah)
target = data[len(data) // 2]

# 3. Implementasi Linear Search
def linear_search(arr, target):
    comparisons = 0  # Menghitung jumlah perbandingan
    for i in range(len(arr)):
        comparisons += 1  # Setiap elemen dibandingkan dengan target
        if arr[i] == target:
            return i, comparisons  # Mengembalikan indeks dan jumlah perbandingan
    return -1, comparisons  # Jika tidak ditemukan

# 4. Implementasi Binary Search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    comparisons = 0

    while left <= right:
        mid = (left + right) // 2
        comparisons += 1  # Setiap kali membandingkan dengan elemen tengah
        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1, comparisons  # Jika tidak ditemukan

# 5. Implementasi Jump Search
def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))  # Ukuran lompatan
    prev = 0
    comparisons = 0

    # Lompat hingga menemukan rentang yang mungkin mengandung target
    while prev < n and arr[min(step, n) - 1] < target:
        comparisons += 1
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1, comparisons

    # Linear search dalam rentang yang ditemukan
    for i in range(prev, min(step, n)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons

    return -1, comparisons  # Jika tidak ditemukan

# 6. Menjalankan eksperimen
linear_result = linear_search(data, target)
binary_result = binary_search(data, target)
jump_result = jump_search(data, target)

# 7. Menampilkan hasil perbandingan jumlah perbandingan yang dilakukan
print("Hasil Eksperimen:")
print(f"Linear Search: {linear_result[1]} perbandingan")
print(f"Binary Search: {binary_result[1]} perbandingan")
print(f"Jump Search: {jump_result[1]} perbandingan")
```

```python
import random
import math
from tabulate import tabulate  # Untuk menampilkan hasil dalam bentuk tabel

# 1. Membuat daftar 100 angka acak dalam rentang 1-1000 dan mengurutkannya
data = sorted(random.sample(range(1, 1001), 100))

# 2. Menentukan angka target (mengambil elemen tengah)
target = data[len(data) // 2]

# 3. Implementasi Linear Search
def linear_search(arr, target):
    comparisons = 0  # Menghitung jumlah perbandingan
    for i in range(len(arr)):
        comparisons += 1  # Setiap elemen dibandingkan dengan target
        if arr[i] == target:
            return i, comparisons  # Mengembalikan indeks dan jumlah perbandingan
    return -1, comparisons  # Jika tidak ditemukan

# 4. Implementasi Binary Search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    comparisons = 0

    while left <= right:
        mid = (left + right) // 2
        comparisons += 1  # Setiap kali membandingkan dengan elemen tengah
        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1, comparisons  # Jika tidak ditemukan

# 5. Implementasi Jump Search
def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))  # Ukuran lompatan
    prev = 0
    comparisons = 0

    # Lompat hingga menemukan rentang yang mungkin mengandung target
    while prev < n and arr[min(step, n) - 1] < target:
        comparisons += 1
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1, comparisons

    # Linear search dalam rentang yang ditemukan
    for i in range(prev, min(step, n)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons

    return -1, comparisons  # Jika tidak ditemukan

# 6. Menjalankan eksperimen
linear_result = linear_search(data, target)
binary_result = binary_search(data, target)
jump_result = jump_search(data, target)

# 7. Menampilkan hasil perbandingan dalam bentuk tabel
table = [
    ["Linear Search", linear_result[1]],
    ["Binary Search", binary_result[1]],
    ["Jump Search", jump_result[1]]
]

print("Hasil Eksperimen Algoritma Searching:")
print(tabulate(table, headers=["Algoritma Pencarian", "Jumlah Perbandingan"], tablefmt="grid"))
```

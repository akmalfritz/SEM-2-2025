```python
# Definisi Node dalam Doubly Linked List
class PageNode:
    def __init__(self, url):
        self.url = url
        self.next = None
        self.prev = None

# Untuk Riwayat
class BrowserHistory:
    def __init__(self):
        self.current = None # riwayat kosong

    # Menambahkan halaman ke riwayat
    def visit_page(self, url):
        new_page = PageNode(url)
        if not self.current:
            self.current = new_page
        else:
            new_page.prev = self.current
            self.current.next = new_page
            self.current = new_page

    # Menampilkan riwayat kunjungan
    def show_history(self):
        if not self.current:
            print("Riwayat kunjungan kosong.")
            return

        print("Riwayat Kunjungan:")
        temp = self.current
        while temp.prev:
            temp = temp.prev
        count = 1
        while temp:
            print(f"{count}. {temp.url}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke halaman sebelumnya, corrected indentation
    def go_back(self):
        if self.current and self.current.prev: # Fixed syntax error: and.selfcurrent.prev
            self.current = self.current.prev
            print(f"Navigasi mundur ke: {self.current.url}")
        else:
            print("Anda berada di halaman pertama.")

    # Navigasi maju ke halaman berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"Navigasi maju ke: {self.current.url}")
        else:
            print("Anda berada di halaman terakhir.")

# Program Utama

# Membuat Objek riwayat
history = BrowserHistory()

# memasukkan jumlah halaman yg dikunjungi
num_pages = int(input("Masukkan jumlah halaman yang ingin dikunjungi: "))

# Loop untuk memasukkan halaman ke dalam doubly linked list
for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page(url)

# Menampilkan riwayat yg telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True: # Changed true to True
    command = input("\n Tekan 'b' untuk navigasi mundur, 'f' untuk navigasi maju, atau 'q' untuk keluar.\nMasukkan perintah: ").lower()
    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'q':
        print("Keluar dari program.")
        break
    else:
        print("Perintah tidak valid. Silakan coba lagi.")
```

```python
# program untuk menampilkan riwayat navigasi folder dalam File Explorer menggunakan Doubly Linked List
class FolderNode:
    def __init__(self, folder_name):
        self.folder_name = folder_name
        self.next = None
        self.prev = None

class FileExplorerHistory:
    def __init__(self):
        self.current = None

    def visit_folder(self, folder_name):
        new_folder = FolderNode(folder_name)
        if not self.current:
            self.current = new_folder
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder

    def show_history(self):
        if not self.current:
            print("\nRiwayat navigasi kosong.")
            return

        print("\nRiwayat Navigasi File Explorer:")
        temp = self.current
        while temp.prev:
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.folder_name}")
            temp = temp.next
            count += 1

    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke: {self.current.folder_name}")
        else:
            print("\nTidak bisa kembali, ini adalah folder pertama!")

    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke: {self.current.folder_name}")
        else:
            print("\nTidak bisa maju, ini adalah folder terakhir!")

history = FileExplorerHistory()
num_folders = int(input("Masukkan jumlah folder yang ingin dinavigasikan: "))

for i in range(1, num_folders + 1):
    folder_name = input(f"Masukkan nama folder ke-{i}: ")
    history.visit_folder(folder_name)
history.show_history()
while True:
    command = input("\nTekan 'b' untuk kembali, 'f' untuk maju, 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print('\nKeluar dari navigasi File Explorer.')
        break
    else:
        print("\nPerintah tidak dikenal.")
```

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang list!")
                return
            temp = temp.next
        if not temp:
            print("Posisi melebihi panjang list!")
            return
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next
        return

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print("Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end="->")
            temp = temp.next
        print("None")

# ======== Program Utama (Main) ========

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))

for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n Pilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```

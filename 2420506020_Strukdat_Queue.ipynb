```python
# Implementasi Queue List
class queueList:
    def __init__(self):
        self.queue = []

    def enqueue(self, data):
        self.queue.append(data)

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None
    def is_empty(self):
        return len(self.queue) == 0
    def size(self):
        return len(self.queue)

# Contoh penggunaan
# Changed 'QueueList' to 'queueList' to match the class definition
queue = queueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue size:", queue.size())
```

```python
# Tugas 1
class TicketQueue:
    def __init__(self, max_size=10):
        self.queue = []  # List untuk menyimpan antrian pelanggan
        self.max_size = max_size  # Kapasitas maksimal antrian

    def is_empty(self):
        return len(self.queue) == 0  # Mengecek apakah antrian kosong

    def size(self):
        return len(self.queue)  # Mengembalikan jumlah pelanggan dalam antrian

    def enqueue(self, ticket_id):
        if '9' in ticket_id:  # Jika pelanggan adalah VIP (mengandung angka 7)
            print(f"Pelanggan {ticket_id} adalah VIP, langsung mendapatkan tiket!")
        elif self.size() < self.max_size:  # Jika antrian belum penuh
            self.queue.append(ticket_id)
            print(f"Pelanggan {ticket_id} ditambahkan ke antrian.")
        else:  # Jika antrian penuh
            print("Antrian penuh! Pelanggan harus menunggu.")

    def dequeue(self):
        if self.is_empty():  # Jika antrian kosong
            print("Tidak ada pelanggan dalam antrian.")
        else:
            served_customer = self.queue.pop(0)  # Menghapus pelanggan pertama dalam antrian
            print(f"Pelanggan {served_customer} telah dilayani.")

    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pelanggan pertama dalam antrian: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pelanggan terakhir dalam antrian: {self.queue[-1]}")

# penggunaan
queue = TicketQueue()
queue.enqueue("Ak20")  # Ditambahkan ke antrian
queue.enqueue("Bn30")  # Ditambahkan ke antrian
queue.enqueue("Xy29")  # Pelanggan VIP langsung dapat tiket
queue.enqueue("Cd40")  # Ditambahkan ke antrian
queue.enqueue("Ef50")  # Ditambahkan ke antrian
queue.enqueue("Xd77")
queue.enqueue("Xd17")
queue.enqueue("Jd77")
queue.front()  # Menampilkan pelanggan pertama
queue.rear()   # Menampilkan pelanggan terakhir
queue.dequeue()  # Melayani pelanggan pertama
queue.size()  # Menampilkan jumlah pelanggan dalam antrian
```

```python
# Tugas 2

class PhotocopyQueue:
    def __init__(self):
        self.queue = []  # List untuk menyimpan antrian pesanan fotokopi

    def is_empty(self):
        return len(self.queue) == 0  # Mengecek apakah antrian kosong

    def size(self):
        return len(self.queue)  # Mengembalikan jumlah pesanan dalam antrian

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:  # Fast Track untuk pesanan kurang dari 10 halaman
            print(f"Pesanan {order_id} dengan {num_pages} halaman langsung diproses tanpa antrian.")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} dengan {num_pages} halaman ditambahkan ke antrian.")

    def dequeue(self):
        if self.is_empty():
            print("Tidak ada pesanan dalam antrian.")
        else:
            order_id, num_pages = self.queue.pop(0)  # Mengambil pesanan pertama dalam antrian
            if num_pages > 100:
                remaining_pages = num_pages - 100
                self.queue.append((order_id, remaining_pages))
                print(f"Pesanan {order_id} diproses 100 halaman, {remaining_pages} halaman tersisa dalam antrian.")
            else:
                print(f"Pesanan {order_id} dengan {num_pages} halaman telah diproses.")

    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pesanan pertama dalam antrian: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pesanan terakhir dalam antrian: {self.queue[-1]}")

# Contoh penggunaan
queue = PhotocopyQueue()
queue.enqueue("Akm20", 5)  # Fast Track, langsung diproses
queue.enqueue("Ssr30", 120)  # Masuk antrian, lebih dari 100 halaman
queue.enqueue("Xpr45", 50)  # Masuk antrian
queue.front()  # Menampilkan pesanan pertama
queue.rear()   # Menampilkan pesanan terakhir
queue.dequeue()  # Melayani pesanan pertama
queue.size()  # Menampilkan jumlah pesanan dalam antrian
```

```python
# Tugas 3

class ParkingQueue:
    def __init__(self, max_size=15):
        self.queue = []  # List untuk menyimpan antrian kendaraan
        self.max_size = max_size  # Kapasitas maksimal parkiran

    def is_empty(self):
        return len(self.queue) == 0  # Mengecek apakah parkiran kosong

    def size(self):
        return len(self.queue)  # Mengembalikan jumlah kendaraan dalam antrian

    def enqueue(self, vehicle_number, is_electric=False):
        if is_electric:  # Kendaraan listrik mendapat prioritas
            self.queue.insert(0, vehicle_number)
            print(f"Kendaraan listrik {vehicle_number} mendapat prioritas masuk parkiran.")
        elif self.size() < self.max_size:  # Jika parkiran belum penuh
            self.queue.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} ditambahkan ke parkiran.")
        else:  # Jika parkiran penuh
            print("Parkiran penuh! Kendaraan harus menunggu.")

    def dequeue(self):
        if self.is_empty():  # Jika parkiran kosong
            print("Tidak ada kendaraan dalam parkiran.")
        else:
            vehicle_out = self.queue.pop(0)  # Mengeluarkan kendaraan pertama dalam antrian
            print(f"Kendaraan {vehicle_out} telah keluar dari parkiran.")

    def front(self):
        if self.is_empty():
            print("Parkiran kosong.")
        else:
            print(f"Kendaraan pertama dalam parkiran: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Parkiran kosong.")
        else:
            print(f"Kendaraan terakhir dalam parkiran: {self.queue[-1]}")

# Contoh penggunaan
parking = ParkingQueue()
parking.enqueue("Be20")  # Ditambahkan ke parkiran
parking.enqueue("Cd30")  # Ditambahkan ke parkiran
parking.enqueue("El45", is_electric=True)  # Kendaraan listrik, mendapat prioritas
parking.front()  # Menampilkan kendaraan pertama
parking.rear()   # Menampilkan kendaraan terakhir
parking.dequeue()  # Mengeluarkan kendaraan pertama
parking.size()  # Menampilkan jumlah kendaraan dalam parkiran
```

```python
# Tugas 4

class CertificateQueue:
    def __init__(self, max_size=20):
        self.regular_queue = []  # List untuk antrian reguler
        self.achiever_queue = []  # List untuk antrian mahasiswa berprestasi
        self.max_size = max_size  # Kapasitas maksimal sebelum dibagi dua jalur

    def is_empty(self):
        return len(self.regular_queue) == 0 and len(self.achiever_queue) == 0

    def size(self):
        return len(self.regular_queue) + len(self.achiever_queue)

    def enqueue(self, student_id, is_achiever=False):
        if is_achiever:  # Mahasiswa berprestasi langsung mendapatkan sertifikat
            print(f"Mahasiswa {student_id} (berprestasi) langsung mendapatkan sertifikat.")
        elif self.size() < self.max_size:  # Jika antrian belum mencapai batas
            self.regular_queue.append(student_id)
            print(f"Mahasiswa {student_id} ditambahkan ke antrian reguler.")
        else:  # Jika antrian penuh, mahasiswa reguler masuk ke jalur terpisah
            self.achiever_queue.append(student_id)
            print(f"Mahasiswa {student_id} ditambahkan ke antrian mahasiswa berprestasi.")

    def dequeue(self):
        if not self.is_empty():
            if self.achiever_queue:
                student_out = self.achiever_queue.pop(0)
                print(f"Mahasiswa {student_out} dari jalur prestasi telah mengambil sertifikat.")
            else:
                student_out = self.regular_queue.pop(0)
                print(f"Mahasiswa {student_out} dari jalur reguler telah mengambil sertifikat.")
        else:
            print("Tidak ada mahasiswa dalam antrian.")

    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            if self.achiever_queue:
                print(f"Mahasiswa pertama dalam jalur prestasi: {self.achiever_queue[0]}")
            elif self.regular_queue:
                print(f"Mahasiswa pertama dalam jalur reguler: {self.regular_queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            if self.regular_queue:
                print(f"Mahasiswa terakhir dalam jalur reguler: {self.regular_queue[-1]}")
            elif self.achiever_queue:
                print(f"Mahasiswa terakhir dalam jalur prestasi: {self.achiever_queue[-1]}")

# Contoh penggunaan
queue = CertificateQueue()
queue.enqueue("T20")  # Ditambahkan ke antrian reguler
queue.enqueue("S30")  # Ditambahkan ke antrian reguler
queue.enqueue("P40", is_achiever=True)  # Mahasiswa berprestasi, langsung dapat sertifikat
queue.front()  # Menampilkan mahasiswa pertama
queue.rear()   # Menampilkan mahasiswa terakhir
queue.dequeue()  # Melayani mahasiswa pertama
queue.size()  # Menampilkan jumlah mahasiswa dalam antrian
```

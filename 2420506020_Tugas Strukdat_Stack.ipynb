```python
# Tugas

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

def move_disk(source, target, source_name, target_name):
    disk = source.pop()
    target.push(disk)
    print(f"Memindahkan cakram {disk} dari {source_name} ke {target_name}")

def hanoi_iterative(n, source, auxiliary, target, moves_limit):
    total_moves = 0
    stacks = {"A": source, "B": auxiliary, "C": target}
    names = list(stacks.keys())

    if n % 2 == 0:
        names[1], names[2] = names[2], names[1]

    while total_moves < moves_limit and target.size() < n:
        for i in range(3):
            from_tower = stacks[names[i]]
            to_tower = stacks[names[(i+1) % 3]]

            if not from_tower.is_empty() and (to_tower.is_empty() or from_tower.peek() < to_tower.peek()):
                move_disk(from_tower, to_tower, names[i], names[(i+1) % 3])
                total_moves += 1
            elif not to_tower.is_empty() and (from_tower.is_empty() or to_tower.peek() < from_tower.peek()):
                move_disk(to_tower, from_tower, names[(i+1) % 3], names[i])
                total_moves += 1

            if total_moves >= moves_limit or target.size() == n:
                break

def main():
    n = 3  # Minimal 3 cakram
    moves_limit = 20  # Dua digit terakhir NPM

    source = Stack()
    auxiliary = Stack()
    target = Stack()

    for i in range(n, 0, -1):
        source.push(i)

    hanoi_iterative(n, source, auxiliary, target, moves_limit)

main()
```

```python
class stack:
    def  __init__(self):
        self.items=[]
    def push(self,item):
        self.items.append(item)
    def pop(self):
        if len(self.items) > 0:
            return self.items.pop()
        else:
            return None

# contoh penggunaan
s = stack()
s.push(1)
s.push(2)
s.push(3)
print(s.items)
s.pop()
print(s.items)
```

```python
class TextEditor:
    def __init__(self):
        self.history = []

    def write(self, text):
        self.history.append(text)

    def undo(self):
        if self.history:
            return f"Undo: {self.history.pop()}"
        else:
            return "No more text to undo"

# Contoh penggunaan
editor = TextEditor()
editor.write("Hello, ")
editor.write("world!")
editor.write("!")
print(editor.undo())
print(editor.undo())
```

```python
from collections import deque

class BrowserHistory:
  def __init__(self):
    self.history = deque()

  def visit(self, url):
    self.history.append(url)
    print(f"Visiting {url}")

  def back(self):
    if self.history:
        print(f"Back to {self.history.pop()}")

    else:
        print("No more pages to go back to")

  def forward(self):
    if self.history:
        print(f"Forward to {self.history.pop()}")

# Contoh Penerapan
browser = BrowserHistory()
browser.visit("www.google.com")
browser.visit("www.facebook.com")
browser.visit("www.twitter.com")
browser.back()
browser.back()
browser.forward()
browser.forward()
```

```python
from queue import LifoQueue

class BackupSystem:
    def __init__(self):
        self.backups = LifoQueue()

    def backup(self, version):
        self.backups.put(version)

    def latest_backup(self):
        if not self.backups.empty():
            return f"Lihat Draft Terakhir {self.backups.queue[-1]}"
        return "Tidak ada backup yang tersedia"

# Contoh penerapan
backup_system = BackupSystem()
backup_system.backup("Ak20@email.com")
backup_system.backup("Ak20@email.com")
backup_system.backup("Ak20@email.com")
print(backup_system.latest_backup())
```

```python
# Lat 1

class Calculator:
    def __init__(self, initial_value):
        self.value = initial_value
        self.history = []

    def add(self, num):
        self.history.append(self.value)
        self.value += num

    def subtract(self, num):
        self.history.append(self.value)
        self.value -= num

    def multiply(self, num):
        self.history.append(self.value)
        self.value *= num

    def divide(self, num):
        if num != 0:
            self.history.append(self.value)
            self.value /= num
        else:
            print("Tidak bisa dibagi dengan nol.")

    def undo(self):
        if self.history:
            self.value = self.history.pop()
            print(f"Undo berhasil. Nilai saat ini: {self.value}")
        else:
            print("Tidak ada operasi untuk di-undo.")

calculator = Calculator(20)

calculator.add(12)
print(f"Setelah penambahan: {calculator.value}")

calculator.subtract(7)
print(f"Setelah pengurangan: {calculator.value}")

calculator.multiply(5)
print(f"Setelah perkalian: {calculator.value}")

calculator.undo()
calculator.undo()
calculator.undo()
```

```python
# Lat 2

from collections import deque

class AppNavigation:
    def __init__(self):
        self.history = deque()

    def visit_feature(self, feature_name):
        self.history.append(feature_name)
        print(f"Mengunjungi fitur: {feature_name}")

    def back(self):
        if len(self.history) > 1:
            self.history.pop()
            print(f"Kembali ke fitur: {self.history[-1]}")
        else:
            print("Tidak ada riwayat sebelumnya!")


def generate_feature_name(birth_month, npm_last_two):
    return birth_month[0].upper() + npm_last_two

app = AppNavigation()


birth_month = "Mei"
npm_last_two = "20"


feature1 = generate_feature_name(birth_month, npm_last_two) + "_Home"
feature2 = generate_feature_name(birth_month, npm_last_two) + "_Pengaturan"
feature3 = generate_feature_name(birth_month, npm_last_two) + "_Profil"

app.visit_feature(feature1)
app.visit_feature(feature2)
app.visit_feature(feature3)

app.back()
app.back()
app.back()
```

```python
# Lat 3

from queue import LifoQueue

class SistemPenyimpananDraftEmail:
    def __init__(self):
        self.drafts = LifoQueue()

    def simpan_draft(self, nama_mahasiswa, npm, konten):
        """Menyimpan draft email ke dalam Stack."""
        subjek = f"{nama_mahasiswa[:2]}{npm[-2:]}"
        draft = {"subjek": subjek, "konten": konten}
        self.drafts.put(draft)
        print(f"Draft email dengan subjek '{subjek}' disimpan.")

    def lihat_draft_terakhir(self):
        """Melihat draft email terakhir yang disimpan."""
        if not self.drafts.empty():
            draft = self.drafts.queue[-1]
            print("\nDraft Email Terakhir:")
            print(f"Subjek: {draft['subjek']}")
            print(f"Konten: {draft['konten']}")
        else:
            print("Tidak ada draft email yang tersimpan.")

sistem_email = SistemPenyimpananDraftEmail()

sistem_email.simpan_draft("Hammam", "2420506018", "Isi email draft 1")
sistem_email.simpan_draft("Salwa", "2420506019", "Isi email draft 2")
sistem_email.simpan_draft("Akmal", "2420506020", "Isi email draft 3")

sistem_email.lihat_draft_terakhir()
```

```python
# Lat 4

class SistemAntreanTiketBioskop:
    def __init__(self):
        self.tickets = []

    def add_ticket(self, bulan_lahir, npm):
        """Menambahkan tiket berdasarkan huruf pertama bulan lahir + dua digit terakhir NPM."""
        nama_film = bulan_lahir[0].upper() + npm[-2:]
        self.tickets.append(nama_film)
        print(f"Tiket untuk film '{nama_film}' telah dibeli.")

    def process_ticket(self):
        """Memproses tiket terakhir dalam antrean (LIFO)."""
        if not self.is_empty():
            return f"Memproses tiket: {self.tickets.pop()}"
        return "Tidak ada tiket yang harus diproses."

    def is_empty(self):
        """Memeriksa apakah Stack kosong."""
        return len(self.tickets) == 0

bioskop = SistemAntreanTiketBioskop()


bioskop.add_ticket("Mei", "2420506020")
bioskop.add_ticket("Mei", "2420506020")
bioskop.add_ticket("Mei", "2420506020")
bioskop.add_ticket("Mei", "2420506020")



print(bioskop.process_ticket())
print(bioskop.process_ticket())
print(bioskop.process_ticket())
print(bioskop.process_ticket())
```
